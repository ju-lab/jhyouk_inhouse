```python
# convert continuous value to categorical bed
# in the case of >= target_value in replication timing

import math

input_file = file("gc50base.mm10.bed")
output_file = file("gc50base_mm10_high.bed","w")

target_value = 50
input_line = input_file.readline().strip()

prev_value = - 999
prev_pos1 = 0
prev_pos2 = 0
prev_chr = 0

peak_start = 0
peak_end = 0
while input_line:
    input_split = input_line.split('\t')
    input_chr = int(input_split[0].replace("chr","").replace("X","23").replace("Y","24"))
    input_pos1 = int(input_split[1])
    input_pos2 = int(input_split[2])
    input_value = float(input_split[3])
    temp_x = 0

    if input_chr > prev_chr:
        if input_value >= target_value:
            peak_start = input_pos1
            peak_end = input_pos2
        else:
            peak_start = 0
            peak_end = 0
    else:
        if prev_value < target_value and input_value < target_value:
            'blank'
        elif prev_value < target_value and input_value >= target_value:
            temp_x = (target_value - prev_value) * (input_pos1 - prev_pos2) / (input_value - prev_value) + prev_pos2
            peak_start = int(math.ceil(temp_x))
        elif prev_value >= target_value and input_value >= target_value:
            peak_end = input_pos2
        else:
            temp_x = (target_value - prev_value) * (input_pos1 - prev_pos2) / (input_value - prev_value) + prev_pos2
            peak_end = int(math.ceil(temp_x))
            output_file.write("%s\t%s\t%s\n" %(input_chr,peak_start,peak_end))
    
    
    prev_value = input_value
    prev_pos1 = input_pos1
    prev_pos2 = input_pos2
    prev_chr = input_chr
    
    input_line = input_file.readline().strip()



output_file.close()
print 'THE END'
```
